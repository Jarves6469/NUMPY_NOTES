program to count the number of each vowel in a string.

```python

def vowel_count(text):
  vowel=['aeiouAEIOU']
  global count
  for i in text:
    if i==vowel:
      count+=i
  return count
s=input("enter sentence :")
vowel_count(s)
print(s)
```

Program to transpose a matrix using nested loop

```python
import numpy as np
a=np.array([[10,20,30],[40,50,60]])
#print(len(a))
rows=len(a)
cols=len(a[0])
# Initialize result matrix with zeros, with dimensions (rows x cols) to hold the transpose
result = np.array([[0,0,],[0,0],[0,0]])
for i in range(rows):
  for j in range(cols):
     result[j][i]=a[i][j]
for k in result:
    print(k)
```

Program to multiply two matrices using nested loops

```python
a=np.array([[1,2],[4,5]])
b=np.array([[3,6],[7,8]])
"""d=a.dot(b)
print(d)"""
"""print(a.ndim)
print(b.ndim)
"""

# More efficient way to achieve the same result
result = np.zeros((len(a), len(b[0])))
for i in range(len(a)):
  for j in range(len(b[0])):
    for k in range(len(b)):
      result[i][j]+=a[i][k]*b[k][j]

for r in result:
  print(r)

```

**SET & METHOD OF SET**

```python
set={1,2,3,4,5,1,2}
set1={8,7,6}
#unordered
print("UNORDERED OR UNIQUE ITEMS :",set)
#MUTABLE
set.add(9)
print("adding new item :",set)
#update(iterable)
set.update(set1)
print("updated new set :",set)
#remove(values)
set.remove(9)# raises KeyError if values is not found
print("remove one items in set :",set)
#discard(values)
set.discard(11)#raises no error if values is not found
print("value not found :",set)
#pop()
set.pop()#remove random element (set unordered)
print("using pop method to remove items :", set)
#clear()
set.clear()
print("using clear method to removes all items from the set :",set)
```

**METHAMETICS OPERATION**

```python
set={1,2,3,4,5,1,2,8}
set1={8,7,6,4}
#union(other set)
set.union(set1)
print("using union method to get common item from other set and returns new set: ",set)
#insersection(other_set)
s=set.intersection(set1)
print("using intersection method to get only common ites from both set :",s)
#diffence(other)
d=set.difference(set1)
print("using difference method to get only difference items from first set :",d)
#symmetric_difference(other)
s_d=set.symmetric_difference(set1)
print("using Symmetric_difference method to get only difference items from both set",s_d)
```

**map(function, iterable, ...)**
** function : The transformation or logic to apply to each element.**
**iterable: One or more collections (list, tuple, etc.) to be processed.**

**Return Value: It returns a map object, which is a lazy iterator. To see the actual values, you must convert it to a collection like a list() or set().**

```python
# squere every number in a list
num=[1,2,3,4]
squered=list(map(lambda x:x**2,num))
print(squered)
```

```python
#Apply existing Python functions like str, int, or abs to entire collections. using map()
x=['1',"2","3"]
integer=list(map(int,x))
integer
```

```python
#sort using map()
x=["32156798"]
sorting=list(map(lambda i:''.join(sorted(i)),x))
sorting
```

**Filter(function,iterable)**

```python
#check EVEN and ODD
l=[1,2,3,1,4,5,6,7,8,9,10]
even=list(filter(lambda x: x%2==0,l))
print(even)
odd=list(filter(lambda x:x%2!=0,l))
print(odd)
```
